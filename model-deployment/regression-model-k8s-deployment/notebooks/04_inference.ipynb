# 04: Inference (API via Minikube)
Calls FastAPI deployment in Minikube using `API_URL` or `http://localhost:30080` via port-forward.

## Inference via Kubernetes (Minikube)

This section calls the FastAPI service deployed to Minikube. Ensure the service is reachable either by port-forwarding (localhost:30080) or using the Minikube IP with NodePort.

- Windows: run scripts/k8s_deploy_minikube.ps1 (includes port-forward)
- Linux/Mac: run scripts/k8s_deploy_minikube.sh

Set API_URL env if not using localhost: http://<minikube-ip>:30080.
